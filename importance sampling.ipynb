```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# unnormalized real distribution : N(z|-1, 1) + N(z|3, 1) * 1.5
def p(z):
    return norm.pdf(z + 1) + norm.pdf(z - 3) * 1.5

def f(z):
    return 0.4 * np.sin(z * 0.2)

# proposal distribution q(z) = N(1, 2^2)
def q(z):
    return norm.pdf(z, 1, 2)
```

```python
Z = np.linspace(-10, 10, 1000)
p_Z = p(Z)
q_Z = q(Z)
f_Z = f(Z)

plt.figure(figsize=[18, 3])
plt.scatter(Z, p_Z, s=2)
plt.scatter(Z, q_Z, s=2)
plt.scatter(Z, f_Z, s=2)
plt.show()
```

```python
L = 10000
# sample from q(z)
samples_Z = (norm.rvs(size=L) * 2) + 1

# p(z^(l))
p_Z_sampled = p(samples_Z)

# q(z^(l))
q_Z_sampled = q(samples_Z)

# f(z^(l))
f_Z_sampled = f(samples_Z)

# E[f]
f_expected = np.mean(p_Z_sampled / q_Z_sampled * f_Z_sampled)
print('E[f] : ', f_expected)
```
